```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# STEP 2 ‚Äî Load the Dataset
titanic = pd.read_csv("Titanic.csv")

# STEP 3 ‚Äî Basic Info
print("üîπ Dataset Information:")
print(titanic.info())

print("\nüîπ Summary Statistics:")
print(titanic.describe())

print("\nüîπ Missing Values:")
print(titanic.isnull().sum())

# STEP 4 ‚Äî Value Counts
print("\nüîπ Survived Count:")
print(titanic['Survived'].value_counts())

print("\nüîπ Passenger Class Count:")
print(titanic['Pclass'].value_counts())

# STEP 5 ‚Äî Visualizations
plt.figure(figsize=(6,4))
sns.countplot(x='Survived', data=titanic, palette='pastel')
plt.title("Survival Count")
plt.show()

plt.figure(figsize=(6,4))
sns.countplot(x='Pclass', hue='Survived', data=titanic, palette='Set2')
plt.title("Survival by Passenger Class")
plt.show()

plt.figure(figsize=(6,4))
sns.histplot(titanic['Age'].dropna(), bins=30, kde=True, color='skyblue')
plt.title("Age Distribution")
plt.show()

plt.figure(figsize=(6,4))
sns.boxplot(x='Survived', y='Age', data=titanic, palette='muted')
plt.title("Age vs Survival")
plt.show()

plt.figure(figsize=(6,4))
sns.barplot(x='Sex', y='Survived', data=titanic, palette='cool')
plt.title("Survival Rate by Gender")
plt.show()

# STEP 6 ‚Äî Correlation Heatmap
plt.figure(figsize=(8,6))
sns.heatmap(titanic.corr(numeric_only=True), annot=True, cmap='Blues')
plt.title("Correlation Heatmap")
plt.show()

# STEP 7 ‚Äî Observations
print("\n‚úÖ Observations:")
print("""
1. More passengers did not survive (0) than those who did (1).
2. Passengers in 1st class had higher survival than 3rd class.
3. Females had a higher survival chance than males.
4. Younger passengers tended to survive more.
5. 'Cabin' and 'Age' columns have missing values.
""")

# STEP 8 ‚Äî Summary
print("\nüìã Summary:")
print("""
The Titanic dataset shows that passenger class, gender, and age
played major roles in determining survival chances.
First-class passengers and females were prioritized during rescue,
while many third-class male passengers did not survive.
""")
```
